```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# Parameters for the normal distribution
mu = 0       # mean
sigma = 1    # standard deviation

# Generate values on the x axis (a range of points)
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 1000)

# Calculate the probability density function (PDF) for x values
y = norm.pdf(x, mu, sigma)

# Create the plot
plt.figure(figsize=(8, 5))  # Optional: specifies the figure size
plt.plot(x, y, label='Normal Distribution', color='blue')

plt.xlim(-4,4)
plt.ylim(0,0.42)

# Optionally, fill under the curve
plt.fill_between(x, y, 0, alpha=0.1, color='blue')

# Add title and labels
# plt.title('Random Variable')
# plt.xlabel('x')
# plt.ylabel('Probability Density')
# plt.legend()

# Display the plot
plt.show()
```
